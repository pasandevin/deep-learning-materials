```python
import numpy as np

data=np.load('dataset/data.npy')
target=np.load('dataset/target.npy')

print(data[:15])
print(target[:15])

print(data.shape)
print(target.shape)
```

```python
from matplotlib import pyplot as plt

plt.plot(data[:2000],label='main')
plt.plot(target[:2000],label='fridge')
plt.legend()
plt.xlabel('time (s)')
plt.xlabel('power consumption (KWh)')
plt.show()
```

```python
print(data.shape)
print(target.shape)
```

```python
new_data=[]
new_target=[]

window_size=1000
stride=10

for i in range(0,60000,stride):
    
    new_data.append(data[i:i+window_size])
    new_target.append(target[i:i+window_size])
```

```python
new_data=np.array(new_data)
new_target=np.array(new_target)
```

```python
print(new_data.shape)
print(new_target.shape)
```

```python
np.save('data',new_data)
np.save('target',new_target)
```
